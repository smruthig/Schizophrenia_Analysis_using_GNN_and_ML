```python
def preprocessing_pipeline(X):
  return X
```

```python
def split_fn(df):
  X = df.drop(['Schizophrenic','Subject'], axis=1)
  y = df['Schizophrenic']

  return X, y
```

```python
def main(df, clf, param):
  # data
  X, y = split_fn(df)
  X = preprocessing_pipeline(X)

  # grid
  gscv = GridSearchCV(clf, param)
  gscv.fit(X, y)
  return gscv
```

```python
clf = XGBClassifier(learning_rate=0.02,  objective='binary:logistic', silent=True, nthread=2)


parameters = {
    'loss': ['log_loss', 'exponential'],
    'max_depth': range (2, 50, 1),
    'n_estimators': range(5, 300, 50)
}
```
